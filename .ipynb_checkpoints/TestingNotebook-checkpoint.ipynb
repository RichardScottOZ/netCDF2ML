### Loading a netCDF file using xarray

```python
file = 'Complete_TAVG_EqualArea.nc'
```

```python
ds = xr.open_dataset(file)
ds
```

```python
ds.temperature
```

```python
temp = ds.temperature

temp1d = temp.isel(time=2015)
```

Doing some coordinate math

```python
x = ds.map_points.longitude.values
y = ds.map_points.latitude.values
```

```python
np.size(x)
```

```python
np.size(y)
```

```python
np.size(temp1d)
```

Now we have all the indicies of the nan's

```python
bad_indicies = np.array(np.where(np.isnan(temp1d.values)))
```

```python
new_temp = np.delete(temp1d.values, bad_indicies)


new_x = np.delete(x, bad_indicies)
new_y = np.delete(y, bad_indicies)

# Quick sanity check that they are all the same
np.size(new_y) == np.size(new_x) == np.size(new_temp)
```

# Simple mapping using MPL

```python
plt.figure(figsize=(7, 6))
plt.title("Projected coordinates of Temperature dataset - Raw Points")
# Plot the bathymetry data locations as black dots
plt.plot(new_x, new_y, ".k", markersize=0.5)
plt.xlabel("Easting")
plt.ylabel("Northing")
plt.gca().set_aspect("equal")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 6))
plt.title('Tricontourf from MPL of the data')
plt.tricontourf(new_x, new_y, new_temp, levels=14)
plt.gca().set_aspect("equal")
plt.xlabel("Easting")
plt.ylabel("Northing")
plt.tight_layout()
```

## Verde Exploration

```python
# We're choosing the latitude of true scale as the mean latitude of our dataset
projection = pyproj.Proj(proj="merc")

proj_coords = projection(new_x, new_y)
```

```python
spacing = 10 / 60
reducer = vd.BlockReduce(np.median, spacing=spacing * 111e3)
filter_coords, filter_bathy = reducer.filter(proj_coords, new_temp)
spline = vd.Spline().fit(filter_coords, filter_bathy)
```
